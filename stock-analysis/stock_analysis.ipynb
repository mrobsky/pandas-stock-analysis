```python
import pandas as pd
import pandas_datareader.data as pdr
import datetime
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
```

```python
start = datetime.datetime(2021,2,8)
end = datetime.datetime(2023,1,18)
yf.pdr_override()
```

```python
cof = pdr.get_data_yahoo("COF", start=start, end=end)
fb = pdr.get_data_yahoo("META", start=start, end=end)
amazon = pdr.get_data_yahoo("AMZN", start=start, end=end)
```

```python
cof['Open'].plot(label = 'COF', figsize=(15,7))
fb['Open'].plot(label='META')
amazon['Open'].plot(label='AMZN')
plt.title('Stock prices for COF, META, AMZN')
plt.ylabel('Price $')
plt.legend()
```

```python
cof['Volume'].plot(label = 'COF', figsize=(15,7))
fb['Volume'].plot(label='META')
amazon['Volume'].plot(label='AMZN')
plt.ylabel('Volume Traded MM')
plt.legend()
```
